```python
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
input_text = 5000
```

```python
m = X.copy()
```

```python
m
```

```python
m=m['title']
```

```python
m
```

```python
m.reset_index(inplace=True)
```

```python
m.reset_index(inplace=True)
```

```python
m['title'][9]
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
corpus = []
for i in range(0, len(m)):
  print(i) ## to check the speed 
  review = re.sub('[^a-zA-Z]', ' ', m['title'][i])
  review = review.lower()
  review = review.split() ## split func because we have to use remove stopwords

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)







```

```python
corpus
```

```python
for words in corpus:
  print(words)
 
```

```python
## One hot encoding of text
```

```python
one = [ one_hot(input_text = words, n=input_text ) for words in corpus ]
one
```

```python
len(one)
```

```python
## Embedding Representation
```

```python
l = 20
embedded_docs = pad_sequences(one,maxlen=l,padding='pre')
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
## creating model 

dim = 40 
model = Sequential()
model.add(Embedding(input_text ,dim,input_length = l ))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam', metrics = ['accuracy'])
model.summary()
```

```python
len(embedded_docs), Y.shape
```

```python
embedded_docs
```

```python
import numpy as np

## for train test split
X_final = np.array(embedded_docs)
Y_final = np.array(Y)
```

```python
X_final
```

```python
X_final.shape , Y_final.shape
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(X_final , Y_final , test_size = 0.33 , random_state = 0) 
```

```python
x_train.shape
```

```python
## Model Training  
```

```python
model.fit(x_train , y_train , validation_data= (x_test, y_test) , epochs = 10 , batch_size = 64)
```

```python
## Adding dropout : to increase val accuray 
```

```python
from tensorflow.keras.layers import Dropout

dim = 40 
model = Sequential()
model.add(Embedding(input_text ,dim,input_length = l ))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam', metrics = ['accuracy'])
model.summary()
```

```python
model.fit(x_train , y_train , validation_data= (x_test, y_test) , epochs = 10 , batch_size = 64)
```

```python
y_pred= model.predict_classes(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
